## **Question 2**
Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
# Select columns 'Y2017' and 'Area'
df = food_df[['Y2017', 'Area']]

# Perform a groupby operation on 'Area'
grouped_df = df.groupby('Area')

# Calculate the sum of 'Y2017' for each Area
sum_df = grouped_df['Y2017'].sum()

# Sort the sum_df DataFrame by 'Y2017' in ascending order
sum_df = sum_df.sort_values()

# Get the 7th lowest sum
seventh_lowest_sum = sum_df.iloc[6]

# Find the Area with the 7th lowest sum
seventh_lowest_sum_area = sum_df.index[sum_df == seventh_lowest_sum][0]

# Print the Area with the 7th lowest sum
print(seventh_lowest_sum_area)
```

**Question 20:** What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python
missing_data_Y2016 = food_df.isnull().sum().loc['Y2014']
percent_missing = round((missing_data_Y2016 * 100 / len(food_df)), 3)

print('The total number of missing data in 2016 = ', missing_data_Y2016)
print('The % of the missing data in 2016 = ', percent_missing,'%')
```
