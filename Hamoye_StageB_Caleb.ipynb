**Question 17 :**
From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the Root Mean Squared error in three D.P?

```python
x = energydata_df['T2'].values.reshape(-1, 1)  # Independent variable
y = energydata_df['T6'].values  # Dependent variable

# Create and fit the linear regression model
model = LinearRegression()
model.fit(x, y)

# Make predictions
y_pred = model.predict(x)

# Calculate the RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))

# Print RMSE rounded to three decimal places
print(f'RMSE: {rmse:.3f}')
```

**Question 18 :**
Remove the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a  random state of 42 (for reproducibility). Normalize the dataset using the MinMaxScaler (Hint: Use the MinMaxScaler fit_transform and transform methods on the train and test set respectively). Run a multiple linear regression using the training set. Answer the following questions:

What is the Mean Absolute Error (in three decimal places) for the  training set?

```python
columns_to_remove = ["date", "lights"]
energydata_df = energydata_df.drop(columns=columns_to_remove)

# Define the features (independent variables) and the target variable
X = energydata_df.drop(columns=["Appliances"])
y = energydata_df["Appliances"]

# Split the data into a training set and a test set (70-30 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalize the dataset using MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train a multiple linear regression model on the training set
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions on the training set
y_train_pred = model.predict(X_train_scaled)

# Calculate the Mean Absolute Error (MAE) for the training set
mae = mean_absolute_error(y_train, y_train_pred)

# Print MAE rounded to three decimal places
print(f"Mean Absolute Error (training set): {mae:.3f}")
```

**Question 19 :**
What is the Root Mean Squared Error (in three decimal places) for the training set?

```python
# Calculate the Mean Squared Error (MSE) for the training set
mse = mean_squared_error(y_train, y_train_pred)

# Calculate the RMSE
rmse = (mse ** 0.5)

# Print RMSE rounded to three decimal places
print(f"Root Mean Squared Error (training set): {rmse:.3f}")
```

**Question 20:**
What is the Mean Absolute Error (in three decimal places) for test set?

```python

# Make predictions on the test set
y_test_pred = model.predict(X_test_scaled)

# Calculate the Mean Absolute Error (MAE) for the test set
mae = mean_absolute_error(y_test, y_test_pred)

# Print MAE rounded to three decimal places
print(f"Mean Absolute Error (test set): {mae:.3f}")
```

**Question 21:**
What is the Root Mean Squared Error (in three decimal places) for test set?

```python
# Calculate the Mean Squared Error (MSE) for the test set
mse = mean_squared_error(y_test, y_test_pred)

# Calculate the RMSE
rmse = (mse ** 0.5)

# Print RMSE rounded to three decimal places
print(f"Root Mean Squared Error (test set): {rmse:.3f}")
```

**Question 24:**
Train a lasso regression model with default value and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
lasso_model = Lasso()

lasso_model.fit(X_train_scaled, y_train)

feature_weights = lasso_model.coef_

# Count the number of features with non-zero feature weights
num_nonzero_features = len(feature_weights[feature_weights != 0])

print(f"Number of features with non-zero feature weights: {num_nonzero_features}")
```

**Question 25:**
What is the new RMSE with the Lasso Regression on the test set?

```python
# Make predictions on the test set using Lasso regression
y_test_pred_lasso = lasso_model.predict(X_test_scaled)

# Calculate the RMSE for Lasso regression on the test set
mse_test_lasso = mean_squared_error(y_test, y_test_pred_lasso)
rmse_test_lasso = (mse_test_lasso ** 0.5)

print(f"RMSE (Lasso Regression on test set): {rmse_test_lasso:.3f}")
```
